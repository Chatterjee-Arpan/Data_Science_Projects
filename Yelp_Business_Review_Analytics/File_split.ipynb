```python
input_file = 'yelp_dataset/yelp_academic_dataset_review.json'
output_prefix = 'split_file_'
num_files = 25 #No. of splits

#Counting total lines in the files
with open(input_file, 'r', encoding='utf-8') as f:
    total_lines = sum(1 for line in f)

lines_per_file = total_lines // num_files #Lines per split file

print(f'Total lines: {total_lines}, Lines per file: {lines_per_file}')

#Splitting into multiple files
with open(input_file, 'r', encoding='utf-8') as f:
    for i in range(num_files):
        output_file = f'{output_prefix}{i+1}.json'

        with open(output_file, 'w', encoding='utf-8') as out_f:
            for _ in range(lines_per_file):
                line = f.readline()
                if not line:
                    break
                out_f.write(line)


print('Splitting completed.')
```
